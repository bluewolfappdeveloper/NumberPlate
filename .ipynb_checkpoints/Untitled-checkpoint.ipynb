```python
import cv2 as cv2
import numpy as np
import time
import pandas as pd
```

```python
path = "GreenParking/0002_02183_b.jpg";
```

```python
def nothing(x):
    pass

def loadImage(path):
    img = cv2.imread(path)
    img = img[20:(img.shape[0]-20), 20:(img.shape[1]-20)]
    return img

def imgShow(img, nameWindow = "hi"): 
    cv2.imshow(nameWindow, img)
    cv2.waitKey(0)

    
def createTrackBar(cv2, n, namedWindow, nameTrackBar, valueTrackBar, todo):
    cv2.namedWindow(namedWindow)
    
    for i in range(0,n):
        cv2.createTrackbar(nameTrackBar[i],namedWindow,0,valueTrackBar[i],nothing)
    
    while(1):
       
        k = cv2.waitKey(1) & 0xFF
        if k == 27:
            break
        
        todo(cv2, namedWindow, nameTrackBar);
        
        
cv2.destroyAllWindows()
# create trackbars for color change
```

```python
img = loadImage(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imgEqualize = cv2.equalizeHist(img)
img = cv2.bilateralFilter(img,30,60,30)
imgShow(img)
```

```python
def tryBilateralFilter(cv2, namedWindow, nameTrackBar):
    pic = cv2.cvtColor(loadImage(path), cv2.COLOR_BGR2GRAY)
    d = cv2.getTrackbarPos(nameTrackBar[0],namedWindow)
    sigmaColor = cv2.getTrackbarPos(nameTrackBar[1],namedWindow)
    sigmaSpace = cv2.getTrackbarPos(nameTrackBar[2],namedWindow)
    pic = cv2.bilateralFilter(pic,d, sigmaColor, sigmaSpace)
    cv2.imshow(namedWindow, pic)    
    
createTrackBar(cv2, 3, "bilateralFilter", ["d", "sigmaColor", "sigmaSpace"], [10, 400, 400], tryBilateralFilter)
```

```python
def preProcessing(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.equalizeHist(img)
    img = cv2.bilateralFilter(img,30,60,30)
    imgShow(img)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
    morph_image = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel,iterations=20)
    imgShow(morph_image)
    sub_morp_image = cv2.subtract(img,morph_image)
    imgShow(sub_morp_image)
    ret,thresh_image = cv2.threshold(sub_morp_image,0,255,cv2.THRESH_OTSU)


    dilated_image = cv2.dilate(thresh_image,kernel,iterations=1)
    return dilated_image;
```

```python
def findPlateNum(img):
    contours, hierarchy = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    contours= sorted(contours, key = cv2.contourArea, reverse = True)[:5]

    img_check = loadImage(path)
    print(img_check)
    
    for contour in contours:
        peri = cv2.arcLength(contour, True)

        approx = cv2.approxPolyDP(contour, 0.08 * peri, True)

        print(peri, len(approx))

        #if (len(approx) ==4):
        (x,y,w,h) = cv2.boundingRect(contour)
        cv2.rectangle(img_check, (x, y), (x + w, y + h), (0, 255, 0), 2)
          #  break
   
        imgShow(img_check)
        time.sleep(3)




```
