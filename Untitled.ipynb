```python
import cv2 as cv2
import numpy as np
import time
import pandas as pd
```

```python
def nothing(x):
    pass

def loadImage(path, isCrop = True):
    img = cv2.imread(path)
    if (isCrop): img = img[20:(img.shape[0]-20), 20:(img.shape[1]-20)]
    return img

def imgShow(img, nameWindow = "hi"): 
    cv2.imshow(nameWindow, img)
    cv2.waitKey(0)

    
def createTrackBar(cv2, n, namedWindow, nameTrackBar, valueTrackBar, todo):
    cv2.namedWindow(namedWindow)
    
    for i in range(0,n):
        cv2.createTrackbar(nameTrackBar[i],namedWindow,0,valueTrackBar[i],nothing)
    
    while(1):
       
        k = cv2.waitKey(1) & 0xFF
        if k == 27:
            break
        
        todo(cv2, namedWindow, nameTrackBar);
        
        
cv2.destroyAllWindows()
# create trackbars for color change
```

```python
def preProcessingV2(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #chuyen anh thanh anh xam
    #can bang sang, giam noise, blur background
    imgEqualize = cv2.equalizeHist(img)
    imgFilter = cv2.bilateralFilter(imgEqualize,9,47,30)

    #tang shape xu dung kernel 5x5, xoa phong
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
    morph_image = cv2.morphologyEx(imgFilter,cv2.MORPH_OPEN,kernel,iterations=8)
    imgSubtract = cv2.subtract(imgFilter,morph_image)

    #dua anh ve trang den, tach biet nen va vat
    ret,thresh_image = cv2.threshold(imgSubtract,0,255,cv2.THRESH_OTSU)
    dilated_image = cv2.dilate(thresh_image,(3,3),iterations=1)
    return dilated_image
```

```python
class PlateInfo:
    def __init__(self, x, y, w, h):
        self.x = x
        self.y = y
        self.w = w
        self.h = h
    def getInfo(self):
        return self.x,self.y,self.w,self.h
```

```python
path = "picPlate/0126_05966_b.jpg";
img = loadImage(path,False)

# imgShow(img)

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


dim = (70, 90)
resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

w, h = resized.shape

resized1 = resized[0:round(2*h/3), :]
resized2 = resized[round(1.5*h/3):, :]

preProcess(resized1)
preProcess(resized2)
```

```python
def preProcess(img):
    imgEqualize = cv2.equalizeHist(img)

    blur = cv2.GaussianBlur(imgEqualize,(3,3),0)
    th2 = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,11,4)

    imgShow(th2)
    
    contours, hierarchy = cv2.findContours(th2, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    contours= sorted(contours, key = cv2.contourArea, reverse = True)[0:10]
    draw = img
    plateInfo = []
    imgShow(img)
    for c in contours:
        x,y,w,h = cv2.boundingRect(c)
        
        if w<=h and x>0 and y > 2 and w >= 3 and h >= 18 :
            
            plateInfo.append(PlateInfo(x,y,w,h))
    
    plateInfo = sorted(plateInfo, key=lambda PlateInfo: PlateInfo.x) 
    
    for i in range(len(plateInfo)):
        if (i == 0 or plateInfo[i].x + plateInfo[i].w > plateInfo[i-1].x + plateInfo[i-1].w):
            x,y,w,h = plateInfo[i].getInfo()
            cv2.rectangle(draw,(x,y),(x+w,y+h),(0,155,0),1)
            imgShow(draw)
    print("Finished");
        
    
```
