```python
import cv2 as cv2
import numpy as np
import time
import pandas as pd
```

```python
def nothing(x):
    pass

def loadImage(path):
    img = cv2.imread(path)
    img = img[20:(img.shape[0]-20), 20:(img.shape[1]-20)]
    return img

def imgShow(img, nameWindow = "hi"): 
    cv2.imshow(nameWindow, img)
    cv2.waitKey(0)

    
def createTrackBar(cv2, n, namedWindow, nameTrackBar, valueTrackBar, todo):
    cv2.namedWindow(namedWindow)
    
    for i in range(0,n):
        cv2.createTrackbar(nameTrackBar[i],namedWindow,0,valueTrackBar[i],nothing)
    
    while(1):
       
        k = cv2.waitKey(1) & 0xFF
        if k == 27:
            break
        
        todo(cv2, namedWindow, nameTrackBar);
        
        
cv2.destroyAllWindows()
# create trackbars for color change
```

def tryMorphologyEx(cv2, namedWindow, nameTrackBar):

    if (MORPH_RECT_Y == 0): MORPH_RECT_Y = 1;

    morph_image = cv2.morphologyEx(pic,cv2.MORPH_OPEN,kernel,iterations=iterations)
    pic = cv2.subtract(pic,morph_image)

createTrackBar(cv2, 3, "morphologyEx", ["MORPH_RECT_X", "MORPH_RECT_Y", "iterations"], [30, 30, 40], tryMorphologyEx)

```python
def preProcessingV2(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    imgEqualize = cv2.equalizeHist(img)
    imgFilter = cv2.bilateralFilter(imgEqualize,9,47,30)

    kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
    morph_image = cv2.morphologyEx(imgFilter,cv2.MORPH_OPEN,kernel,iterations=8)
    imgSubtract = cv2.subtract(imgFilter,morph_image)

    ret,thresh_image = cv2.threshold(imgSubtract,0,255,cv2.THRESH_OTSU)
    dilated_image = cv2.dilate(thresh_image,(3,3),iterations=1)
    return dilated_image
```

```python
def findPlateNum(img):
    contours, hierarchy = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    contours= sorted(contours, key = cv2.contourArea, reverse = True)[:2]

    img_check = loadImage(path);
#     (x,y,w,h) = cv2.boundingRect(contours[0])
#     cv2.rectangle(imgResult, (x, y), (x + w, y + h), (0, 255, 0), 2)
#     return imgResult
    isChoose = False; edge = [0,0]; dem = -1;
    for contour in contours:
        peri = cv2.arcLength(contour, True)

        approx = cv2.approxPolyDP(contour, 0.1 * peri, True)

        dem+=1;
        edge[dem] = len(approx);
        #print(peri, len(approx))

        if (len(approx) ==4 and peri > 250):
            isChoose = True
            (x,y,w,h) = cv2.boundingRect(contour)
            cv2.rectangle(img_check, (x, y), (x + w, y + h), (0, 255, 0), 2)
            break

   
    if (isChoose == False):
        if (edge[1] > edge[0] and cv2.arcLength(contours[1], True) > 200):
            (x,y,w,h) = cv2.boundingRect(contours[1])
            cv2.rectangle(img_check, (x, y), (x + w, y + h), (0, 255, 0), 2)
        else:
            (x,y,w,h) = cv2.boundingRect(contours[0])
            cv2.rectangle(img_check, (x, y), (x + w, y + h), (0, 255, 0), 2)
    return img_check;
```
